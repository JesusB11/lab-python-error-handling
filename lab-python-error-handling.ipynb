# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
def get_customer_orders(inventory):
    customer_orders = 0
    order_products = []
     
    while customer_orders == 0:
        num_orders= int(input("Please indicate if you're placing 1 or 2 customer orders: ").strip())
        try:
            if num_orders == 2:
                customer_orders= 2
                break
            elif num_orders == 1:
                continue_ordering = input("Are you sure you want to place just one order? (yes/no): ").lower()
                while continue_ordering not in ["yes", "no"]: # entrada valida
                    continue_ordering = input("Please enter a valid answer. Do you want to continue ordering? (yes/no): ").lower()
                if continue_ordering == "no":
                    customer_orders = 2
                elif continue_ordering == "yes":
                    customer_orders= 1
        except:
            print("Please enter a number between 1 and 2") 
    if customer_orders > 0:
            try:
                while int(len(order_products)) < int(customer_orders):
                    for i in range(int(customer_orders)):                             
                            product_name = input(f"Enter the name of product #{i+1} that goes in the order: ").lower()
                            if product_name in inventory.keys(): 
                                order_products.append(product_name) 
                            else:
                                print(f"'{product_name}' is not in the inventory. Please enter a valid product")
            except KeyboardInterrupt:
                print("\nOrder process interrupted by user during product input. Order creation stopped.")
    
    return (customer_orders, int(num_orders), set(order_products))

def initialize_inventory(products):
                        
    """ This is desing to get input for the inventory"""
    inventory = {product: int(input(f"Enter the initial amount for product: {product} in the inventory")) for product in products} 

    return inventory 
 
def update_inventory(order_products, inventory):
    """This function updates the inventory based on customer orders"""
    print(order_products,"44")
    for item in order_products:
        if item in inventory:
            inventory[item] = inventory[item] -1 
    
    updated_inventory= {key: value for key, value in inventory.items() if value > 0}
    return updated_inventory
    
def calculate_order_statistics(num_orders, inventory):
    
    sum_inventory_final=len(inventory.keys())
    total_products_percentage=(num_orders/sum_inventory_final*100) 
    order_statistics = (total_products_percentage, num_orders)
    return order_statistics

def print_order_statistics(order_statistics):
    
    print("Order Statistics:")
    print("Total products ordered: ", order_statistics[1]) 
    print(f"Percentage of products ordered: {order_statistics[0]:.2f}%") 

    return order_statistics

def print_updated_inventory(updated_inventory):
    print("Updated inventory:")
    for product,value in updated_inventory.items():
        print(f"{product}: {value}")

def calculate_total_prize(order_products):
    prices=0
    while prices <= 0:
        try:
            prices= sum([float(input(f"Enter the value of {value}: ")) for value in order_products])
            if prices > 0:
                break      
            else:
                print("Price must be greater than 0")          
        except ValueError:
            print("Please enter a valid price for the product(s).")
    return prices


products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
customer_orders, num_orders, order_products= get_customer_orders(inventory)
updated_inventory= update_inventory(order_products, inventory)
prices=calculate_total_prize(order_products)
order_statistics= calculate_order_statistics(num_orders, inventory)


print_order_statistics(order_statistics)

print_updated_inventory(updated_inventory)
print(f"Total Price {prices}")
````
